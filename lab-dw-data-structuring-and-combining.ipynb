```python
# Import Pandas library
import pandas as pd
```

```python
# Load data list from url
url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
insurance_df1 = pd.read_csv(url1)
```

```python
insurance_df1
```

```python
# Load data list from url
url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
insurance_df2 = pd.read_csv(url2)
```

```python
insurance_df2
```

```python
# Load data list from url
url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
insurance_df3 = pd.read_csv(url3)
```

```python
insurance_df3
```

```python
insurance_df1.columns = [column.lower() for column in insurance_df1.columns]
insurance_df2.columns = [column.lower() for column in insurance_df2.columns]
insurance_df3.columns = [column.lower() for column in insurance_df3.columns]
```

```python
insurance_df1.columns = [column.replace(' ','_') for column in insurance_df1.columns]
insurance_df2.columns = [column.replace(' ','_') for column in insurance_df2.columns]
insurance_df3.columns = [column.replace(' ','_') for column in insurance_df3.columns]
```

```python
insurance_df1.rename(columns={"st": "state"}, inplace=True)
insurance_df2.rename(columns={"st": "state"}, inplace=True)
```

```python
insurance_df1
```

```python
insurance_df2
```

```python
insurance_df3
```

```python
insurance_df1["number_of_open_complaints"].value_counts()
```

```python
insurance_df1["number_of_open_complaints"].unique()
```

```python
insurance_df1["number_of_open_complaints"]=insurance_df1["number_of_open_complaints"].fillna('1/0/00')
```

```python
insurance_df1["number_of_open_complaints"].unique()
```

```python
insurance_df1["number_of_open_complaints"]=insurance_df1["number_of_open_complaints"].astype(str)
```

```python
insurance_df1["number_of_open_complaints"].unique()
```

```python
insurance_df1["number_of_open_complaints"]=insurance_df1["number_of_open_complaints"].apply(lambda noc: noc.split("/")[1])
```

```python
insurance_df1["number_of_open_complaints"]=insurance_df1["number_of_open_complaints"].astype(int)
```

```python
insurance_df2["number_of_open_complaints"].value_counts()
```

```python
insurance_df2["number_of_open_complaints"]=insurance_df2["number_of_open_complaints"].astype(str)
```

```python
insurance_df2["number_of_open_complaints"]=insurance_df2["number_of_open_complaints"].apply(lambda noc: noc.split("/")[1])
```

```python
insurance_df2["number_of_open_complaints"]=insurance_df2["number_of_open_complaints"].astype(int)
```

```python
insurance_df1.set_index('customer', inplace=True)
insurance_df2.set_index('customer', inplace=True)
insurance_df3.set_index('customer', inplace=True)
```

```python
insurance_df_all = pd.concat([insurance_df1, insurance_df2, insurance_df3], axis=0)
```

```python
insurance_df_all
```

```python
insurance_df_all.nunique()
```

```python
insurance_df_all["gender"].unique()
```

```python
insurance_df_all["gender"].value_counts()
```

```python
# Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female",
# which need to be standardized, for example, to "M" and "F".
insurance_df_all["gender"]=insurance_df_all["gender"].apply(lambda gen: "M" if gen=="Male" else "F" if gen=="female" else "F" if gen=="Femal" else gen)
```

```python
insurance_df_all["gender"].value_counts()
```

```python
insurance_df_all["state"].value_counts()
```

```python
insurance_df_all["state"]=insurance_df_all["state"].apply(lambda st: "Washington" if st=="WA" else "Arizona" if st=="AZ" else "California" if st=="Cali" else st)
```

```python
insurance_df_all["state"].value_counts()
```

```python
insurance_df_all["education"].value_counts()
```

```python
insurance_df_all["education"]=insurance_df_all["education"].apply(lambda edu: "Bachelor" if edu=="Bachelors" else edu)
```

```python
insurance_df_all["education"].value_counts()
```

```python
insurance_df_all["customer_lifetime_value"]=insurance_df_all["customer_lifetime_value"].apply(lambda clv: str(clv))
```

```python
insurance_df_all["customer_lifetime_value"]=insurance_df_all["customer_lifetime_value"].apply(lambda clv: clv.replace("%","") if "%" in clv else clv)
```

```python
insurance_df_all["customer_lifetime_value"]=insurance_df_all["customer_lifetime_value"].apply(lambda clv: float(clv))
```

```python
insurance_df_all["customer_lifetime_value"].value_counts()
```

```python
insurance_df_all["vehicle_class"].value_counts()
```

```python
insurance_df_all["vehicle_class"]=insurance_df_all["vehicle_class"].apply(lambda vc: "Luxury" if vc=="Sports Car" else "Luxury" if vc=="Luxury SUV" else "Luxury" if vc=="Luxury Car" else vc)
```
